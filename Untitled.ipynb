```python
def longestCommonPrefix(strs):
    """
    :type strs: List[str]
    :rtype: str
    """
    #   check if strs just have only one element
    if len(strs) == 1:
        return strs[0]
    # find shortest word
    shortest_word = strs[0]
    len_shortest_word = len(strs[0])
    for idx, word in enumerate(strs[1:]):
        if len(word) < len_shortest_word:
            shortest_word = word
            len_shortest_word = len(word)
    #  
    for i in range(len(shortest_word)):
        prefix_shortest = shortest_word[:i + 1]
        list_prefix_check = []
        for word in strs:
            if word == shortest_word:
                continue
            else:
                if prefix_shortest != word[: i +1]:
                    return shortest_word[:i]
    return shortest_word     
    
        
```

```python
list_tmp = [["flower","flow","flight"], 
            ["dog","racecar","car"]]
            
for tmp in list_tmp:
    # print(tmp)
    prefix = longestCommonPrefix(tmp)
    print("prefix: ", prefix)
```

```python
def myAtoi(s):
    """

    """
    is_negative = False
    existing_signed = False
    #
    number_char = ''
    number = 0
    # signed idx < number idx
    signed_idx = 0
    number_idx = 0
    for idx, char in enumerate(s):
        if char == ' ':
            signed_idx += 1
            number_idx += 1
            continue
        elif char.isalpha() or char == '.':
            break
        elif char == '-' or char == '+':
            signed_idx = idx
            if signed_idx > number_idx:
                break
            else:
                if char == '-':
                    is_negative = True
                existing_signed = True
        else:    # char is number
            number_idx = idx
            number_char += char
    if len(number_char) < 1:
        number = 0
    else:
        number = int(number_char)
    if is_negative:
        number *= -1
    if number > 2**31-1:
        number = 2**31
    elif number < -2**31:
        number = -2**31
    return number
```

```python
test_cases = ['3.14', '42', ' -042', "1337c0d3", '0-1', "words and 987"] # 
for test_case in test_cases:
    print(myAtoi(test_case))
```

```python
def calSquareNumbers(n):
    total_square_numbers = 0
    if n < 10:
        total_square_numbers = n**2
    while (n / 10 != 0):
        number = n % 10
        total_square_numbers += number**2
        n = int(n / 10)
    return total_square_numbers
```
