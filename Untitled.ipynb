```python
def longestCommonPrefix(strs):
    """
    :type strs: List[str]
    :rtype: str
    """
    #   check if strs just have only one element
    if len(strs) == 1:
        return strs[0]
    # find shortest word
    shortest_word = strs[0]
    len_shortest_word = len(strs[0])
    for idx, word in enumerate(strs[1:]):
        if len(word) < len_shortest_word:
            shortest_word = word
            len_shortest_word = len(word)
    #  
    for i in range(len(shortest_word)):
        prefix_shortest = shortest_word[:i + 1]
        list_prefix_check = []
        for word in strs:
            if word == shortest_word:
                continue
            else:
                if prefix_shortest != word[: i +1]:
                    return shortest_word[:i]
    return shortest_word     
    
        
```

```python
def myAtoi(s):
    """

    """
    is_negative = False
    existing_signed = False
    #
    number_char = ''
    number = 0
    # signed idx < number idx
    signed_idx = 0
    number_idx = 0
    for idx, char in enumerate(s):
        if char == ' ':
            signed_idx += 1
            number_idx += 1
            continue
        elif char.isalpha() or char == '.':
            break
        elif char == '-' or char == '+':
            signed_idx = idx
            if signed_idx > number_idx:
                break
            else:
                if char == '-':
                    is_negative = True
                existing_signed = True
        else:    # char is number
            number_idx = idx
            number_char += char
    if len(number_char) < 1:
        number = 0
    else:
        number = int(number_char)
    if is_negative:
        number *= -1
    if number > 2**31-1:
        number = 2**31
    elif number < -2**31:
        number = -2**31
    return number
```

```python
test_cases = ['3.14', '42', ' -042', "1337c0d3", '0-1', "words and 987"] # 
for test_case in test_cases:
    print(myAtoi(test_case))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Gi s bn c thm d liu demand quantity cho tng route
transport_data = {
    ('TC', 'Plant01-DC01-CZ01'): {'cost': 150000, 'demand': 50000},
    ('TC', 'Plant01-DC01-CZ02'): {'cost': 200000, 'demand': 50000},
    ('TC', 'Plant02-DC02-CZ02'): {'cost': 150000, 'demand': 50000},
    ('TC', 'Plant02-DC02-CZ03'): {'cost': 200000, 'demand': 50000},
    
    ('MC', 'Plant01-DC01-CZ01'): {'cost': 150000, 'demand': 50000},
    ('MC', 'Plant01-DC01-CZ02'): {'cost': 120000, 'demand': 30000},
    ('MC', 'Plant01-DC02-CZ02'): {'cost': 245000, 'demand': 40834},
    ('MC', 'Plant01-DC02-CZ03'): {'cost': 204166, 'demand': 29166}, 
    ('MC', 'Plant02-DC02-CZ02'): {'cost': 87500, 'demand': 29166},
    ('MC', 'Plant02-DC02-CZ03'): {'cost': 83333, 'demand': 20833},
}

# Tch d liu t dictionary
factories_customers = list(transport_data.keys())
costs = [data['cost'] for data in transport_data.values()]
demands = [data['demand'] for data in transport_data.values()]

# To danh sch cc nhn cho mi cp route
x_labels = [f'{factory}-{customer}' for factory, customer in factories_customers]
x_pos = np.arange(len(x_labels))

# Thit lp  rng v v tr ca cc ct
bar_width = 0.4
x_pos_demand = x_pos - bar_width / 2
x_pos_cost = x_pos + bar_width / 2

# Phn loi cc nh my  gn mu
demand_colors = ['#FFCA00']

cost_colors = ['#8DD28F']

plt.figure(figsize=(14, 8))

# V biu  ct cho demand v cost
bars_demand = plt.bar(x_pos_demand, demands, width=bar_width, color=demand_colors, label='Demand Quantity')
bars_cost = plt.bar(x_pos_cost, costs, width=bar_width, color=cost_colors, label='Transportation Cost')

# t tn cho trc v tiu 
plt.xlabel('Network flow From Plant to Customer', fontsize=18)
plt.ylabel('Path costs', fontsize=18)
plt.title('Demand Quantity and Transportation Costs by Factory and Customer', fontsize=22)
plt.legend()

# Thm tn cho cc ct trn trc honh
plt.xticks(x_pos, x_labels, rotation=45, ha='right')

# t gi tr ln trn u mi ct
for bar in bars_demand:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.0f}', ha='center', va='bottom', fontsize=8)

for bar in bars_cost:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.0f}', ha='center', va='bottom', fontsize=8)

# Hin th biu 
plt.tight_layout()
plt.savefig('demand_cost_plot.png', dpi=300)  # Lu vi  phn gii 300 dpi
plt.show()
```

```python
def calSquareNumbers(n):
    total_square_numbers = 0
    if n < 10:
        total_square_numbers = n**2
    while (n / 10 != 0):
        number = n % 10
        total_square_numbers += number**2
        n = int(n / 10)
    return total_square_numbers
```
